```python
import numpy as np
import matplotlib.pyplot as plt

# Definir las frecuencias y amplitudes de las seales
frecuencia_1 = 1
frecuencia_2 = 1
t = np.linspace(0, 2 * np.pi, 1000)

amplitud = 5  # Puedes ajustar esto segn tus necesidades

# Definir los factores de frecuencia para las 8 relaciones
relaciones_frecuencia = [(1, 1), (1, 2), (1, 3), (2, 3), (3, 4), (3, 5), (4, 5), (5, 6)]

# Configurar la figura con 3 filas y 3 columnas para las 8 grficas
fig, axs = plt.subplots(3, 3, figsize=(12, 12))

for i, (frec_1, frec_2) in enumerate(relaciones_frecuencia):

    # Calcular las seales usando las relaciones de frecuencia
    signal_1 = amplitud * np.sin(2 * np.pi * frec_1 * t)
    signal_2 = amplitud * np.sin(2 * np.pi * frec_2 * t)

    # Graficar en el subplot correspondiente
    row, col = divmod(i, 3)
    axs[row, col].plot(signal_1, signal_2)
    axs[row, col].set_title(f'{frec_1}:{frec_2}')
t = np.linspace(0,  np.pi/300, 1000)
# Ajustar el diseo de las subgrficas
axs[2, 2].plot(t, 5*np.sin(600*t), label='10 Vpp sin(600t)')
axs[2, 2].set_title('10 Vpp sin(600t)')
axs[2, 2].legend()

plt.tight_layout()
plt.show()
```
