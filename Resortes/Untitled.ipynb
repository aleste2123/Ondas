```python
import pandas as pd
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
import numpy as np
nombre_archivo_excel = 'u.xlsx'
```

```python
m = np.array([0.37, 0.25,0.13, 0.1,0.03])
sime = np.array([1.245, 1.25, 1.26,1.25, 1.25])
anti = np.array([0.788,0.9,1.04,1.09,1.27])
pul = np.array([1.258, 2.523, 5.78,9.63,1.247])


w_s = 2*np.pi/sime
print(w_s)
w_a =  2*np.pi/anti
print(w_a)
w_pul = 2*np.pi/pul
print(w_pul)
```

```python
plt.plot(m,w_s, label= "Simétrico")
plt.plot(m,w_a, label= "Antisimétrico")
plt.plot(m,w_pul, label= "Pulsación")
plt.xlabel("Altura [m]")
plt.ylabel("Frecuencia [rad/s]")
plt.legend()
```

```python
val = np.sum(w_s)/5
print(val)

g = val**2*385.57*10/(40*(200+104.2/2)+100*0.1)
print(g)
```

```python
def ajuste(h, w0, c):
    return (w0**2+c*2*h**2)**0.5

res_1, con_1 = curve_fit(ajuste, m, w_a, p0=[10,100])

perr = np.sqrt(np.diag(con_1))
print(perr[1])
print(res_1[0])
print(res_1[1]*4*37.56e-3)

plt.plot(m,ajuste(m,*res_1))
plt.errorbar(m,w_a,yerr=0.05,fmt='o')
plt.xlabel("Altura [m]")
plt.ylabel("w [rad/s]")
```
