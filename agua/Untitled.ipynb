```python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
```

```python
def cargar_imagen(ruta_imagen, limites_recorte):
    # Carga la imagen
    with Image.open(ruta_imagen) as img:
        # Recorta la imagen
        img_recortada = recortar_imagen(img, limites_recorte)
        # Convierte la imagen a escala de grises
        img_recortada_gris = img_recortada.convert('L')
        # Convierte la imagen recortada a un array numpy
        imagen_array = np.array(img_recortada_gris)
        # Obtiene las dimensiones de la imagen recortada en píxeles
        ancho_pixeles, alto_pixeles = img_recortada_gris.size
        # Convierte las dimensiones de píxeles a centímetros
        ancho_cm = ancho_pixeles * 0.8/100
        alto_cm = alto_pixeles * 0.8/100
        return imagen_array, ancho_cm, alto_cm

def recortar_imagen(imagen, limites_recorte):
    """
    Recorta la imagen utilizando los límites especificados y devuelve la imagen recortada.
    Los limites_recorte deben ser una tupla (izquierda, superior, derecha, inferior).
    """
    return imagen.crop(limites_recorte)

def mostrar_grafica_imagen(info):
    # Muestra la imagen recortada
    imagen_array, ancho_cm, alto_cm = info
    plt.imshow(imagen_array, cmap='gray', extent=(0, ancho_cm, 0, alto_cm))
    plt.title("Imagen Recortada")
    plt.xlabel("Ancho (cm)")
    plt.ylabel("Alto (cm)")
    plt.colorbar(label='Intensidad de Grises')
    plt.show()

def graficar_intensidad(info, orientacion_eje_x):
    # Determina la orientación del eje x
    imagen_array, ancho_cm, alto_cm = info
    fig, axs = plt.subplots(1, 2, figsize=(10, 5))
    im1 = axs[0].imshow(imagen_array, cmap='gray', extent=(0, ancho_cm, 0, alto_cm))
    axs[0].set_title("Imagen Recortada")
    axs[0].set_xlabel("Ancho (cm)")
    axs[0].set_ylabel("Alto (cm)")
    plt.colorbar(im1, ax=axs[0], label='Intensidad de Grises')
    if orientacion_eje_x == 'fila':
        intensidad = np.mean(imagen_array, axis=0)
        x = np.linspace(0, ancho_cm, len(intensidad))
        axs[1].plot(x, intensidad)
        axs[1].set_title("Intensidad de Grises")
        axs[1].set_xlabel("Eje x [cm]")
        axs[1].set_ylabel("Intensidad Promedio")
        axs[1].scatter(4,130)
    elif orientacion_eje_x == 'columna':
        intensidad = np.mean(imagen_array, axis=1)
        y = np.linspace(0, alto_cm, len(intensidad))
        axs[1].plot(y, intensidad)
        axs[1].set_title("Intensidad de Grises")
        axs[1].set_xlabel("Eje y [cm]")
        axs[1].set_ylabel("Intensidad Promedio")
    plt.subplots_adjust(wspace=0.5)
        

    
    
```

```python
# Ruta de la imagen y límites de recorte
ruta_imagen = 'Captura de Pantalla 2024-04-07 a la(s) 22.02.20.png'
limites_recorte = (400, 0, 700, 500) 

# Carga de la imagen y obtención de dimensiones
info = cargar_imagen(ruta_imagen, limites_recorte)
# Muestra la gráfica de la imagen recortada


# Grafica la intensidad con orientación del eje x en la columna de la mitad de la imagen
graficar_intensidad(info, orientacion_eje_x='columna')
print(64)
```
